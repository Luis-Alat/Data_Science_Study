## Prompting Principles

### __Principle 1: Write clear and specific instructions__

*Tactic 1:* Use delimiters to clearly indicate distinct parts of the input
Delimiters can be anything like: ```, """, < >, <tag> </tag>

    This avoid prompting injections, for example, avoid executing possible instructions given to the model that could cause a confusion about the task asked to chatgpt and the text/data supplied as input to analize

*Tactic 2:* Ask for a structured output

*Tactic 3:* Ask the model to check whether conditions are satisfied. If the task assumption is not satisfied, indicate the expected output in that case or edge cases respectively.

*Tactic 4:* Tactic 4: "Few-shot" prompting

    Give successful examples of completing tasks. Then ask model to performance the task

*Tactic 1*: Specify the steps required to complete a task
*Tactic 2*: Instruct the model to work out its own solution before rushing to a conclusion

* Analyze where the results does give what you want
* Clarify instructions, give more time to think
* Repat until get successful results

* Idea
* Implementation (code/prompt)
* Experimenta results

### ChatBot

There are 3 types of roles in ChatGPT

1. User: The user role is played by the person who is interacting with ChatGPT. As a user, you can ask questions or make statements, and ChatGPT will provide responses based on the input it receives.

2. Assistant: The assistant role is played by ChatGPT itself. As an assistant, ChatGPT's goal is to provide helpful and informative responses to the user's input. ChatGPT is designed to understand natural language input and provide accurate and relevant information based on its training data.

3. System: The system role is played by the underlying software and infrastructure that powers ChatGPT. This includes things like servers, databases, and programming languages. The system's main goal is to ensure that ChatGPT runs smoothly and is able to provide timely and accurate responses to user input.

The intention of these roles is to facilitate a smooth and effective conversation between the user and ChatGPT. The user can input natural language questions and statements, and ChatGPT, acting as an assistant, can provide accurate and relevant responses. The system is there to support the assistant role and ensure that the conversation runs smoothly and without interruption.

````python
# Multiples messages to chatGPT

messages =  [  
{'role':'system', 'content':'You are an assistant that speaks like Shakespeare.'},    
{'role':'user', 'content':'tell me a joke'},   
{'role':'assistant', 'content':'Why did the chicken cross the road'},   
{'role':'user', 'content':'I don\'t know'}  ]

response = gpt.get_completion_from_messages(messages, temperature=1)
print(response)
````

````python
# Virtual Assistent

def collect_messages(_):
    prompt = inp.value_input
    inp.value = ''
    context.append({'role':'user', 'content':f"{prompt}"})
    response = gpt.get_completion_from_messages(context) 
    context.append({'role':'assistant', 'content':f"{response}"})
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600, style={'background-color': '#F6F6F6'})))
 
    return pn.Column(*panels)
````
